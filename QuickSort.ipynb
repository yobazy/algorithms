```python
# implement quicksort

def quickSort(arr):
    length = len(arr)
    
    if length <= 1:
        return(arr)
    else:    
        piv = arr.pop()
    
    higher = []
    lower = []
    
    for item in arr:
        if item > piv:
            higher.append(item)
        else:
            lower.append(item)
            
    return quickSort(lower) + [piv] + quickSort(higher)

arr_unsort = [5, 3, 1, 6, 4]
print(arr_unsort)
arr_sort = quickSort(arr_unsort)
print(arr_sort)
```

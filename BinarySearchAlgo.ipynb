Binary Search Algorithm

```python
#Performs iterative binary search to find the position of an integer in a given, sorted, list.

def binarySearch(inpArr, item):
    first = 0
    last = len(inpArr) - 1
    while first <= last:
        i = int((first+last)/2)
        if inpArr[i]==item:
            return ' found at position %d %d %d' %(item,i,inpArr[i])
        elif inpArr[i]>item:
            last = i - 1 
        elif inpArr[i]<item:
            first = i + 1
        else:
            return ' not found in the list'.format(item=item)
            
A = [1,2,3,4,5,6,7]
x = binarySearch(A,3)
print(x)
```
